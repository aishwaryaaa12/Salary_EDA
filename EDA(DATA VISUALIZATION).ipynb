# Subplots

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,15))

plt.subplot(2,3,1)
sb.countplot(A.Type)

plt.subplot(2,3,2)
sb.countplot(A.Origin)

plt.subplot(2,3,3)
sb.countplot(A.AirBags)

plt.subplot(2,3,4)
sb.distplot(A.Price)

plt.subplot(2,3,5)
sb.distplot(A.Horsepower)

plt.subplot(2,3,6)
sb.distplot(A.Weight)
```

```python
plt.figure(figsize=(40,70))                  # using loops
x = 1

for i in A.columns:
    if(A[i].dtypes == "object"):
        plt.subplot(10,3,x)
        sb.countplot(A[i])
        x = x + 1
    else:
        plt.subplot(10,3,x)
        sb.distplot(A[i])
        x = x + 1
```

# Scatterplot and Box plot

```python
X = A.drop(labels=["id","Make","Model","Max.Price","Min.Price","Price"],axis=1)
Y = A[["Price"]]
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
plt.figure(figsize=(40,60))
a = 1
for i in X.columns:
    if(X[i].dtypes == "object"):
        plt.subplot(8,3,a)
        sb.boxplot(X[i],Y.Price)
        a = a + 1
    else:
        plt.subplot(8,3,a)
        sb.scatterplot(X[i],Y.Price)
        a = a + 1
```
